# Using tractography function in pipeline to generate structure tensors

Import the main analysis module and tractography module.  Use the analysis module to download a raw brain of interest, then use the tractography function "nii_to_tiff_stack" to save a .nii as a TIFF stack.

```python
import tractography as tract
```

```python
import analysis3 as a3
```

```python
output = a3.get_raw_brain("s275", "userToken.pem", resolution = 5, save = True)
```

```python
tract.nii_to_tiff_stack("s275_raw.nii", "s275")
```

```python
data = tract.tiff_stack_to_array("s275_TIFFs/")
```

### From there, run the generate_FSL_structure_tensor code to save a copy of the FSL structure tensor.
Note that the value can also be directly used from the function.

```python
output = tract.generate_FSL_structure_tensor(data, "s275")
```
